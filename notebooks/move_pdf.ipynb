````python
```
This script moves all PDF files from subdirectories to the main directory. It also removes empty subdirectories after moving the PDF files.
```


import os
import shutil

def move_pdfs_to_main_dir(main_dir):
    # Walk through all subdirectories and files in the main directory
    for root, _, files in os.walk(main_dir):
        for file in files:
            if file.lower().endswith('.pdf'):
                source_path = os.path.join(root, file)
                destination_path = os.path.join(main_dir, file)
                
                # Check if the file already exists in the destination
                if os.path.exists(destination_path):
                    print(f"File {file} already exists in {main_dir}. Skipping.")
                else:
                    shutil.move(source_path, destination_path)
                    print(f"Moved {file} to {main_dir}")
    # Remove empty subdirectories
    for root, dirs, _ in os.walk(main_dir, topdown=False):
        for dir in dirs:
            dir_path = os.path.join(root, dir)
            if not os.listdir(dir_path):
                os.rmdir(dir_path)
                print(f"Removed empty directory: {dir_path}")

# Replace 'path_to_main_directory' with the path to your main directory
move_pdfs_to_main_dir('data/AWS')
````
