```python
# process all PDFs in the folder

import os
import json
import pandas as pd
from pdfminer.high_level import extract_text
import spacy


```

```python
# Load spaCy model for Named Entity Recognition (NER)
nlp = spacy.load("en_core_web_sm")

```

```python
# Define paths
pdf_dir = "C:/Users/email/Desktop/DS_Science_Youth_Extension/2504_Science_Youth/data/AWS"
output_json = "processed_papers.json"

```

```python
# Function to clean text
def clean_text(text):
    text = text.replace("\n", " ")  # Remove new lines
    text = " ".join(text.split())  # Remove extra spaces
    return text

```

```python
# Function to extract entities using spaCy
def extract_entities(text):
    doc = nlp(text)
    entities = {ent.label_: ent.text for ent in doc.ents}
    return entities

```

```python
# Process all PDFs
processed_data = []
for pdf_file in os.listdir(pdf_dir):
    if pdf_file.endswith(".pdf"):
        pdf_path = os.path.join(pdf_dir, pdf_file)
        try:
            text = extract_text(pdf_path)
            cleaned_text = clean_text(text)
            entities = extract_entities(cleaned_text)

            processed_data.append({
                "file_name": pdf_file,
                "text": cleaned_text[:5000],  # Limit text size to avoid memory issues
                "entities": entities
            })

            print(f"Processed: {pdf_file}")

        except Exception as e:
            print(f"Error processing {pdf_file}: {e}")

```

```python
# Save results to JSON
with open(output_json, "w", encoding="utf-8") as f:
    json.dump(processed_data, f, indent=4)

```

```python
# Optional: Save to CSV
df = pd.DataFrame(processed_data)
df.to_csv("processed_papers.csv", index=False)

print("Processing complete! Data saved as JSON and CSV.")
```
